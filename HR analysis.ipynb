```python
## Data cleaning 
```

```python
df.isnull().sum()
```

```python
# removing unecessary columns 


columns_to_drop = ["EmployeeCount", "Over18"]  
df_cleaned = df.drop(columns=columns_to_drop)
```

```python
df_cleaned 
```

```python
# elimanting abudant enteries 

df_cleaned = df.drop_duplicates()
df_cleaned
```

```python
# renaming the column 
df_cleaned = df_cleaned.rename(columns={"Attrition": "EmployeeAttrition"})
df_cleaned
```

```python
#Sanatizing appropriate columns
```

```python
# Strip Whitespace
for col in ['EmployeeAttrition', 'BusinessTravel', 'Department', 'EducationField', 'Gender', 'JobRole', 'MaritalStatus', 'Over18', 'OverTime']:
    df_cleaned[col] = df_cleaned[col].astype(str).str.strip()

# Lowercase
for col in ['EmployeeAttrition', 'BusinessTravel', 'Department', 'EducationField', 'Gender', 'JobRole', 'MaritalStatus', 'Over18', 'OverTime']:
    df_cleaned[col] = df_cleaned[col].astype(str).str.lower()
    
# Handle Categorical Integers
df_cleaned['Education'] = df_cleaned['Education'].astype(object) 

# Replace Unknown Values (optional)
for col in ['EmployeeAttrition', 'BusinessTravel', 'Department', 'EducationField', 'Gender', 'JobRole', 'MaritalStatus', 'Over18', 'OverTime']:
    df_cleaned[col] = df_cleaned[col].astype(str).str.replace('?', 'unknown', regex=False)

df_cleaned[col]
```

```python
# converting categorical into numerical
df_cleaned['Education'].replace(['Under-Graduate', 'Diploma'], [0, 1], inplace=True)
df_cleaned['Education']
```

```python
# using lable encoding to convert categorical into numerical
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df_cleaned['Encoded_Department'] = le.fit_transform(df['Department'])

df_cleaned['Encoded_Department']
```

```python
# Additional steps 
```

```python
#  distribution of employees across different departments. 

department_counts = df_cleaned['Department'].value_counts()


colors = ['skyblue', 'salmon', 'lightgreen', 'orange', 'purple']

plt.bar(department_counts.index, department_counts.values, color=colors)
plt.xlabel('Department')
plt.ylabel('Number of Employees')
plt.title('Employee Distribution by Department')
plt.show()
```

```python
#Attrition by Years at Company

attrition_rate = df_cleaned.groupby('YearsAtCompany')['EmployeeAttrition'].value_counts(normalize=True).unstack()
sns.lineplot(data=attrition_rate['yes'], marker='o', color='#FF5733')  # Use a hex color code


attrition_rate['yes'].plot(kind='line', marker='o')
plt.title('Attrition Rate by Years at Company')
plt.xlabel('Years at Company')
plt.ylabel('Attrition Rate')
plt.show()
```

```python
#Salary vs. Attrition


sns.set_palette("husl") # or 'Set2', 'Paired', 'colorblind', 'muted', 'bright'


sns.boxplot(data=df_cleaned, x='EmployeeAttrition', y='MonthlyIncome', showmeans=True, meanprops={"marker": "o",
                       "markerfacecolor": "white", 
                       "markeredgecolor": "black",
                      "markersize": "10"}) 
plt.title('Monthly Income by Attrition Status', fontsize=14)
plt.xlabel('Employee Attrition', fontsize=12)
plt.ylabel('Monthly Income', fontsize=12)
plt.xticks(rotation=45) #Rotate the xlabels for better readability
plt.show()
```

```python
#Relationship Satisfaction vs. Attrition

sns.violinplot(data=df_cleaned, x='EmployeeAttrition', y='RelationshipSatisfaction')
plt.title('Relationship Satisfaction by Attrition Status')
plt.show()
```

```python
#Employee Attrition Overview

sns.set_palette("husl") # or 'Set2', 'Paired', 'colorblind', 'muted', 'bright'

# Create the count plot
sns.countplot(data=df_cleaned, x='EmployeeAttrition')
plt.title('Employee Attrition Distribution', fontsize=14)
plt.xlabel('Employee Attrition', fontsize=12)
plt.ylabel('Count', fontsize=12)

# Show the plot
plt.show()
```

```python
# Employee Tenure vs. Monthly Income by Attrition 

sns.lineplot(data=df_cleaned, x='YearsAtCompany', y='MonthlyIncome', hue='EmployeeAttrition')
plt.title('Employee Tenure vs. Monthly Income by Attrition', fontsize=14)
plt.show()
```

```python
#Job Satisfaction vs. Monthly Income 

sns.set_palette("husl") # or 'Set2', 'Paired', 'colorblind', 'muted', 'bright'

# Create the regression plot with custom scatterplot and line colors
sns.regplot(x='JobSatisfaction', y='MonthlyIncome', data=df_cleaned, scatter_kws={'color': 'skyblue'}, line_kws={'color': 'darkorange'})

plt.title('Job Satisfaction vs. Monthly Income', fontsize=14)
plt.show()

```
